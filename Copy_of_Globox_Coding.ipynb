Using Coalesce Function to add 0 to null values in Spent column

```python
query = ''' SELECT *, COALESCE (a.spent, 0) AS spent
FROM users AS u
LEFT JOIN activity AS a
ON u.id = a.uid
LEFT JOIN groups AS g
ON u.id=g.uid;'''
read_query (query)
```

Using Max/Min functions to determine dates of study

```python
query = ''' SELECT min(dt) AS start_date, max(dt) AS end_date
FROM users AS u
LEFT JOIN activity AS a
ON u.id = a.uid
LEFT JOIN groups AS g
ON u.id=g.uid;'''
read_query (query)
```

Determining number of users in Control and Treatment Groups using CASE Statement

```python
query = ''' SELECT cluster, COUNT (DISTINCT (uid))
FROM (SELECT CASE WHEN "group" = 'A' then 'Control' else 'experimental' end AS cluster, uid
FROM groups as g) AS B
GROUP BY cluster'''
read_query(query)
```

Determining Conversion rate of all users
using avg function and nested CASE statement


```python
query = '''SELECT (avg (convert)*100) AS conversion_rate
FROM
  (SELECT distinct g.uid, CASE WHEN a.spent IS null THEN 0 ELSE 1 END AS convert
FROM groups AS g
LEFT JOIN activity AS a
ON g.uid= a.uid) AS X;'''
read_query(query)
```

Conversion Rate for Control and Experimental Group

```python
query=''' SELECT cluster, (avg (convert)*100) AS conversion_rate
FROM
  (SELECT distinct g.uid, CASE WHEN g.group = 'A' THEN 'control'
   ELSE 'test' END AS cluster,
  CASE WHEN a.spent IS null THEN 0 ELSE 1 END AS convert
FROM groups AS g
LEFT JOIN activity AS a
ON g.uid= a.uid) AS X
GROUP BY cluster
ORDER BY conversion_rate;
'''
read_query(query)
```

```python
query=''' SELECT cluster, SUM(convert) AS Number_success
FROM
  (SELECT distinct g.uid, CASE WHEN g.group = 'A' THEN 'control'
   ELSE 'test' END AS cluster,
  CASE WHEN a.spent IS null THEN 0 ELSE 1 END AS convert
FROM groups AS g
LEFT JOIN activity AS a
ON g.uid= a.uid) AS X
GROUP BY cluster;
'''
read_query(query)
```

Average Amount Spent per user for control and treatment group

```python
query= '''SELECT cluster, SUM(result)/COUNT(DISTINCT(X.uid)) AS dollars_spent_by_each_user
FROM
 (SELECT DISTINCT(g.uid), CASE WHEN g.group = 'A' THEN 'control' ELSE 'test' END
 AS cluster,
 (COALESCE(a.spent,0)) AS result
 FROM groups AS g
 LEFT JOIN activity AS a
 ON g.uid=a.uid) AS X
 GROUP BY cluster;
'''
read_query(query)
```

Exporting Relevant Data into csv
